```python
import numpy as np
from environment import Environment, State, Action
from cryptocurrency import Crypto
import time

from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
""" Here, we'll buy 10 times in a row and see what happens. 
Note that no rewards occur unless a sale is made. Thus, there won't be any here. 
"""

NUM_ITER = 135000

States = []
Rewards = []

env = Environment.with_csv_fetcher()

S_t = State.default()
for t in range(NUM_ITER):
    States.append(S_t)
    (S_t, R_t) = env.step(S_t, Action.buy)
    Rewards.append(R_t)

Profit_At_States = list(map(lambda x: x.total_balance, States))

fig, ax = plt.subplots(2)
ax[0].plot(list(range(NUM_ITER)), Rewards, label="Rewards earned at $t$")
ax[0].set_ylabel("Rewards")
ax[0].set_ylim([-0.005, 0.005])
ax[1].plot(list(range(NUM_ITER)), Profit_At_States, label="Current balance at $t$")
ax[1].set_ylabel("Total balance (USD)")
ax[1].set_ylim([100000 - 25, 100000 + 25])
```

```python
""" Here, we'll buy once, then sell 9 times in a row and see what happens. 
Note that no rewards occur unless a sale is made. Thus, we won't get any rewards until t > 1
"""

States = []
Rewards = []

S_t = State.default()
States.append(S_t)
(S_t, R_t) = Environment.step(S_t, Action.buy)
Rewards.append(R_t)
for t in range(9):
    States.append(S_t)
    time.sleep(10)
    (S_t, R_t) = Environment.step(S_t, Action.sell)
    Rewards.append(R_t)

Profit_At_States = list(map(lambda x: x.total_balance, States))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(list(range(10)), Rewards, label="Rewards earned at $t$")
ax[0].set_ylabel("Rewards")
ax[0].set_ylim([-0.01, 0.022])
ax[1].plot(list(range(10)), Profit_At_States, label="Current balance at $t$")
ax[1].set_ylabel("Total balance (USD)")
ax[1].set_ylim([100000 - 0.05, 100000 + 0.05])
```

```python
""" Here, we'll hold 10 times in a row and see what happens. 
Note that no rewards occur unless a sale is made, and we will expect the balance to not change at all
since by default, we hold no cryptocurrency. 
"""

States = []
Rewards = []


S_t = State.default()
for t in range(10):
    States.append(S_t)
    time.sleep(10)
    (S_t, R_t) = Environment.step(S_t, Action.sell)
    Rewards.append(R_t)

Profit_At_States = list(map(lambda x: x.total_balance, States))

fig, ax = plt.subplots(2)
ax[0].plot(list(range(10)), Rewards, label="Rewards earned at $t$")
ax[0].set_ylabel("Rewards")
ax[0].set_ylim([-0.005, 0.005])
ax[1].plot(list(range(10)), Profit_At_States, label="Current balance at $t$")
ax[1].set_ylabel("Total balance (USD)")
ax[1].set_ylim([99999, 100001])
```
